```python
import pandas as pd    
optional_list = [[("H","latency","max",100), ("O","operational_cost","max",900)],[("H","availability","min",99),("O","reliability","min",89.9)],[("H","min_storage","min",1024),("O","licensing_cost","max",500)],[("H","latency","max",110),("O","licensing_cost","max",700)],[("H","availability","min",80),("O","min_storage","min",1024)],[("H","reliability","min",85),("O","min_storage","min",1024)],[("H","licensing_cost","max",400),("O","operational_cost","max",500)],[("H","latency","max",100),("O","operational_cost","max",400)],[("H","licensing_cost","max",500),("O","operational_cost","max",700)],[("H","min_cpu","min",2),("O","min_ram","min",4)],[("H","latency","max",120),("O","operational_cost","max",550)],[("H","min_cpu","min",2),("O","availability","min",85)]]
inc_path = "C:\\Users\\umroot\\Desktop\\Concordia\\Masters Thesis\\Relevant_INC\\UR"
def getRelevantINC(inc_path,optional_list) :
    #print(optional_list)
    relevant_inc = []
    # this list append all optional preferences
    keys_to_extract = [' inc']
    df = pd.read_csv(inc_path)
    # Convert the DataFrame into a list of dictionaries, one for each row (from csv to dictionary)
    list_of_dicts = df.to_dict(orient='records')
    #print(list_of_dicts)
    column_mapping = {"operational_cost":" oc","licensing_cost":" lc","availability":" a","operation_name":" name","bandwidth":" b","latency":" l","min_ram":" mr","min_cpu":" mc","reliability":" r","min_storage":" ms"}
    for opt in optional_list:
        preference,data_property,condition,value = opt
        # Define keys to extract
        keys_to_extract.append(column_mapping.get(data_property))
   # Create a new list that only contains the user preferences for each dictionary
    inc_list = [{key: d[key] for key in keys_to_extract} for d in list_of_dicts]
    #print(inc_list)
    ranked_incs = []
    for inc in inc_list:
        priority_value = 0
        #match = True  # Assume a match until proven otherwise
        for opt in optional_list:
            preference,data_property,condition,value = opt
            data_property = column_mapping.get(data_property)
            if condition == "max" and data_property in inc and (inc[data_property] <= value):
                priority_value += 100 if preference == "H" else 1
            elif condition == "min" and data_property in inc and (inc[data_property] >= value):
                priority_value += 100 if preference == "H" else 1
            elif condition == "exactly" and data_property in inc and (inc[data_property] == value):
                priority_value += 100 if preference == "H" else 1
        ranked_incs.append((inc[" inc"].replace('"^^anyURI','').replace('"','').strip(),priority_value))
    #print("This is before ranking URIs of INCs")
    #print(ranked_incs)
    # Rank/Sort the ranked_incs based on the priority values in descending order
    sorted_ranked_incs = sorted(ranked_incs, key=lambda x: x[1], reverse=True)
    # Create a new list containing only the URI values of INCs from the sorted list of tuples
    ranked_incs_uris = [ranked_incs[0] for ranked_incs in sorted_ranked_incs]
    print("This is the ranked URIs of INCs based on their preferences.")
    #print(ranked_incs_uris)
    return ranked_incs_uris
#getRelevantINC(inc_path,optional_list[1])    
for i in range(len(optional_list)):
    print("The user preference UR"+str(i)+" has been sent.")
    inc_location = getRelevantINC((inc_path+str(i)+".csv"),optional_list[i])
    print(inc_location)
```
